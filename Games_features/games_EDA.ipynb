```python
lista = df['ReleaseDate'] #extract year from realease date

year = []
for i in lista:
    year.append(re.findall(r'\d\d\d\d', str(i)))

df['year'] = year
df
```

```python

lista = df['year'] #clean characters [] from 'year'
year = []
for i in lista:
    year.append(re.sub(r'\W', "", str(i)))

df['year'] = year
df
```
